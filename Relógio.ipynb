## Pseudocódigo

```TEXT 
SETUP:
    UnidadeSeg = 0
    DezenaSeg = 0
    UnidadeMin = 0
    DezenaMin = 0
    UnidadeHora12 = 2
    UnidadeHora24 = 0
    DezenaHora12 = 1
    DezenaHora24 = 0

LOOP:
    SE CHAVECONFIGHORA FOR 1:
        PULA PRA -CONFIGHORA
    SE CHAVECONFIGMIN FOR 1:
        PULA PRA -CONFIGMIN
    SE CHAVECONFIGSEG FOR 1:
        PULA PRA -CONFIGSEG
    SE CHAVE2412 FOR 1:
        PULA PRA -24
    PULA PRA -12
    
-TEMPO:
    UnidadeSeg ++
    SE UNIDADESEG FOR MAIOR QUE 9
        UnidadeSeg = 0
        DezenaSeg ++
    SE DezenaSeg FOR MAIOR QUE 5
        DezanaSeg = 0
        UnidadeMinu ++
    SE UnidadeMinu FOR MAIOR QUE 9
        UnidadeMinu = 0
        DezenaMin ++
    SE DezenaMin FOR MAIOR QUE 5
        DezenaMin = 0
        UnidadeHora12 ++
        UnidadeHora24 ++
    SE UnidadeHora12 FOR MAIOR 2
        SE DezenaHora12 FOR MAIOR QUE 0
            UnidadeHora12 = 1
            DezanaHora12 = 0
        SE UnidadeHora12 FOR MAIOR QUE 9
            UnidadeHora12 = 0
            DezenaHora12 ++    
    SE UnidadeHora24 FOR MAIOR 3
        SE DezenaHora24 FOR MAIOR QUE 1
            UnidadeHora24 = 0
            DezanaHora24 ++
        SE UnidadeHora24 FOR MAIOR QUE 9
            UnidadeHora24 = 0
            DezenaHora24 ++
    SE DezenaHora24 FOR MAIOR QUE 2
        DezanaHora24 = 0
       
       
-24
    ENVIA PARA 7SEG
        UnidadeSeg
        DezenaSeg
        UnidadeMin
        DezenaMin
        UnidadeHora24
        DezenaHora24
    PULA PARA -TEMPO

-12:
    ENVIA PARA 7SEG
        UnidadeSeg
        DezenaSeg
        UnidadeMin
        DezenaMin
        UnidadeHora12
        DezenaHora12
    PULA PARA -TEMPO

-CONFIGHORA:
    SE BOTAO1 APERTADO
        UnidadeHora12 ++
        UnidadeHora24 ++
    SE BOTAO2 APERTADO
        DezenaHora24 ++
        DezenaHora12 ++


-CONFIGMIN:
    SE BOTAO1 APERTADO
        UnidadeMin ++
    SE BOTAO2 APERTADO
        DezenaMin ++


-CONFIGSEG:
    SE BOTAO1 APERTADO
        UnidadeSeg ++
    SE BOTAO2 APERTADO
        DezenaSeg ++
```

## COMANDOS ASSEMBLY

```
- add Reg1, Reg2, Reg3 (adiciona valor do Reg1 com o valor de Reg3 e salva o resultado no Reg3)

- addi Reg1, imediato (adiciona valor do registrador com o imediato salva o resultado no Reg1)

- load Reg, End (move o que esta em Reg para o Endereço)

- jmp imediato (pula uma determinada quantidade de linhas, determinado pelo imediato)

- cmpi Reg1, imediado, Reg2 (compara o valor do Reg1 com um imediato e salva em Reg2)

- jle Reg, imediato (Pula a quantidade de linhas determinado pelo imediato, se o valor dentro do Reg for menor ou igual a zero)

- rid End, Reg (Lê as informações de um endereço para o Reg)
```

## Instruções

|   OPCODE   |   Reg1   |   Reg2   |   Reg3   |   Imediato   |
|:--------:|:--------:|:--------:|:--------:|:--------:|
|5 bits|5 bits|5 bits|5 bits|5 bits|

